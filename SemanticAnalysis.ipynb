```python
import numpy as np

emotionlabels= ['anger', 'disgust', 'fear', 'joy', 'neutral', 'sadness', 'surprise']
isbn= []

def calculateMeanEmotionScores(predictions):
    perEmotionScore= {label:[] for label in emotionlabels}
    for prediction in predictions:
        sortedPrediction= sorted(prediction, key=lambda x: x['label'])
        for index, label in enumerate(emotionlabels):
            perEmotionScore[label].append(sortedPrediction[index]['score'])
    return {label: np.mean(scores) for label, scores in perEmotionScore.items()}
```

```python
emotionScores= {label: [] for label in emotionlabels}

for i in range(1):
    isbn.append(books['isbn13'][i])
    sentence= books['description'][i].split(".")
    predictions= classifier(sentence)
    meanScores= calculateMeanEmotionScores(predictions)
    for label in emotionlabels:
        emotionScores[label].append(meanScores[label])
```

```python
from tqdm import tqdm

emotionLabels= ['anger', 'disgust', 'fear', 'joy', 'neutral', 'sadness', 'surprise']
isbn= []
emotionScores= {label: [] for label in emotionlabels}

for i in tqdm(range(len(books))):
    isbn.append(books['isbn13'][i])
    sentences= books['description'][i].split(".")
    predictions= classifier(sentences)
    meanScores= calculateMeanEmotionScores(predictions)
    for label in emotionLabels:
        emotionScores[label].append(meanScores[label])
```

```python
emotionDf= pd.DataFrame(emotionScores)
emotionDf['isbn13']=isbn
```
